```python
bd_veiculos = []
bd_usuario = []
id_usuario = 0
siglas_estados = ["AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO", "MA",
                  "MT", "MS", "MG", "PA", "PB", "PR", "PE", "PI", "RJ", "RN",
                  "RS", "RO", "RR", "SC", "SP", "SE", "TO"]
```

```python
def cadastro_veiculo ():
  placa = input("Digite a placa do seu veículo: ")
  print("Agora informe a localização do veículo: \nUtilize a sigla para indicar o estado")
  estado_veiculo = input("\n Estado: ").upper()
  cidade_veiculo = input("\n Cidade: ").title()
  print("Categoria de aluguel \n1-Básico \n2-Intermediário \n3-Luxo")
  categoria_aluguel = int(input("Informe a categoria de aluguel: "))
  valor_diaria = 0
  locado = -1

  if categoria_aluguel == 1:
    categoria_aluguel = "Básico"
    valor_diaria = 150
  elif categoria_aluguel == 2:
    categoria_aluguel = "Intermediário"
    valor_diaria = 250
  elif categoria_aluguel == 3:
    categoria_aluguel = "Luxo"
    valor_diaria = 400
  else:
    print("Opção de aluguel inválido")

  if estado_veiculo not in siglas_estados:
    print("\nDigite apenas a sigla do estado! Ex.: SP")
    estado_veiculo = input("Digite novamente: ").upper()

  veiculo = [placa, estado_veiculo, cidade_veiculo, categoria_aluguel, valor_diaria, locado]

  bd_veiculos.append(veiculo)

  print(f"\n\nDados cadastrados com sucesso confira: \nPlaca: {veiculo[0]} \nLocalização: {veiculo[1]} - {veiculo[2]} \nCategoria: {veiculo[3]} \nValor diária: {veiculo[4]} \nStatus locação: {veiculo[5]}")
  print("\n--Atualização Banco de Dados--")
 ## for i, veiculo in enumerate(bd_veiculos):
  ##  print(f"Nº registro {i}: {veiculo}")
```

```python
def cadastro_cliente():
  nome = input("Digite seu nome: ").title()
  sobrenome = input("Digite seu sobrenome: ").title()

  global id_usuario
  id_usuario += 1
  cliente = [id_usuario, nome, sobrenome]
  bd_usuario.append(cliente)

  print(f"\nDados cadastrados com sucesso confira: \nUsuário: {nome} {sobrenome} - ID {cliente[0]}")
  print("\n--Atualização Banco de Dados--")
  ##for i, cliente in enumerate(bd_usuario):
    ##print(f"Nº registro {i}: {cliente}")
```

```python
def veiculo_estado():
  estado_procura = input("Digite a sigla do estado que deseja mapear os veículos: ").upper()

  if estado_procura not in siglas_estados:
    print("\nDigite a sigla de um estado válido!")
    return
  veiculos_achados = []
  for estado_cadastrado in bd_veiculos:
    if estado_cadastrado[1] == estado_procura:
      veiculos_achados.append(estado_cadastrado)

  if veiculos_achados:
    print(f"\nForam encontrados {len(veiculos_achados)} veiculos cadastrados no {estado_procura}:\n")
    tipo_visualizacao = int(input("Deseja ver quais veículos? \n1- Disponíveis \n2- Indisponíveis\n"))

    if tipo_visualizacao == 1:
      for i, veiculo in enumerate(veiculos_achados):
        if veiculo[5] == -1:
          print(f"Registro {i}: {veiculo}")
        else:
          print("\nNão há registros!")
    elif tipo_visualizacao == 2:
      for i, veiculo in enumerate(veiculos_achados):
        if veiculo[5] != -1:
          print(f"Registro {i}: {veiculo}")
        else:
          print("\nNão há registros!")
  else:
    print(f"Não foram encontrados registros em {estado_procura}")
```

```python
def listar_cadastros_veiculos():
  print(f"\nVeículos cadastrados: {len(bd_veiculos)}\n")
  if bd_veiculos:
    for i, veiculos in enumerate(bd_veiculos):
      print(f"Registro nº {i}: nPlaca: {veiculos[0]} \nLocalização: {veiculos[1]} - {veiculos[2]} \nCategoria: {veiculos[3]} \nValor diária: {veiculos[4]} \nStatus locação: {veiculos[5]}")
      print("------------------------------")
  else:
    print("Não há veículo cadastrado!")
```

```python
def visualizar_cliente():
  tipo_busca = int(input("\nDeseja fazer a busca por: \n1-Nome \n2-Sobrenome \n"))

  clientes_achados = []
  if tipo_busca == 1:
    nome_procurado = input("\nDigite o nome a ser buscado: ").title()
    for cliente in bd_usuario:
      if cliente[1]== nome_procurado:
        clientes_achados.append(cliente)
    if clientes_achados:
      print(f"\nRegistros com o nome {nome_procurado}: {len(clientes_achados)}")
      for i, cliente in enumerate(clientes_achados):
        print(f"Nº {i}: {cliente}")
    else:
      print("\nNão há registros com esse nome")

  elif tipo_busca == 2:
    sobrenome_procurado = input("\nDigite o sobrenome a ser buscado: ").title()
    for cliente in bd_usuario:
      if cliente[2]== sobrenome_procurado:
        clientes_achados.append(cliente)
    if clientes_achados:
      print(f"\nRegistros com o nome {sobrenome_procurado}: {len(clientes_achados)}")
      for i, cliente in enumerate(clientes_achados):
        print(f"Nº {i}: {cliente}")
    else:
      print("\nNão há registros com esse sobrenome")
  else:
    print("\nDigite um número válido")
    return
```

```python
def listar_cadastro_usuarios():
  print(f"\nClientes cadastrados: {len(bd_usuario)}\n")
  if bd_usuario:
    for i, cliente in enumerate(bd_usuario):
      print(f"Usuário nº {i}: \nID: {cliente[0]}\nNome: {cliente[1]} \nSobrenome: {cliente[2]}")
      print("------------------------------")
  else:
    print("Não há usuário cadastrado!")
```

```python
def alugar_veiculo():
  placa = input("\nDigite a placa do veículo que deseja alugar: ")

  for veiculo in bd_veiculos:
    if veiculo[0]==placa:
      if veiculo[5] == -1:
        print(f"\nVeículo {placa} disponível para locação!")
        print(f"Confirme os dados do veículo que deseja alugar: \nPlaca: {veiculo[0]} \nLocalização: {veiculo[1]} - {veiculo[2]} \nCategoria: {veiculo[3]} \nValor diária: {veiculo[4]}")
        confirma = input("Dados estão corretos? (sim/nao): ").lower()

        if confirma == "sim":
          id_usuario = int(input("Digite o ID do cliente que deseja alugar: "))

          for cliente in bd_usuario:
            if cliente[0] == id_usuario:
              print(f"\nConfirme os dados do usuário: \nID: {cliente[0]} \nNome: {cliente[1]} \nSobrenome: {cliente[2]}")
              confirmar = input("\nDados estão corretos? (sim/nao)").lower()

              if confirmar == "sim":
                veiculo[5] = 0
                print(f"\nVeículo {placa} alugado com sucesso para o usuário {id_usuario}")
                print(f"Confira atualização: {veiculo}")
              elif confirmar == "nao":
                novo = input("\Deseja inserir um novo ID? (sim/nao)")

                if novo == "sim":
                  novo_id = int(input("Digite o novo ID: "))
                  id_usuario = novo_id
                else:
                  return
              else:
                print("Digite opção válida!")
                return
            else:
              print("Cliente não encontrado!")
        elif confirma == "nao":
          nova_placa = input("\nDeseja inserir uma nova placa? (sim/nao)").lower()

          if nova_placa == "sim":
            placa = input("Digite a placa de um veículo: ")
            return
          else:
            return
      else:
        print("\nVeículo não encontrado!")
```

```python
def devolucao_carro():
  placa = input("Digite a plca do veículo que deseja devolver: ")

  for veiculo in bd_veiculos:
    if veiculo[0] == placa:
      if veiculo[5] == 0:
        print("\nVeículo localizado!")
        dias = int(input("Digite quantos dias ficou alugado: "))
        preco = dias * veiculo[4]
        print(f"Preço total do aluguel: {preco}")
        confirmado = input("Confirmar devolução e valor? (sim/nao)").lower()

        if confirmado == "sim":
          veiculo[5] = -1
        elif confirmado == "nao":
          return
        else:
          print("valor invalido")
      else:
        print("Veículo não está alugado")
    else:
      print("Veículo não encontrado")
```

```python
def menu ():
   while True:
    print("\n\n***SISTEMA RENT A CAR***")
    print("Insira o número da opção desejada: ")
    print("1- Criar um registro")
    print("2- Listar registros")
    print("3- Buscar registro")
    print("4- Locação de veículo")
    print("5- Sair")

    opcao_menu = int(input("Opção: "))

    if opcao_menu == 1:
      print("***CRIAR REGISTRO***\n")
      opcao = int(input("Deseja criar registro: \n1-Veículo \n2-Cliente \n3-Sair\n"))
      if opcao == 1:
        cadastro_veiculo()
      elif opcao == 2:
        cadastro_cliente()
      else:
        return
    elif opcao_menu == 2:
      print("***LISTAR REGISTROS***\n")
      opcaoDois = int(input("Deseja listar registros: \n1-Veículos \n2-Clientes \n3-Sair\n"))
      if opcaoDois == 1:
        listar_cadastros_veiculos()
      elif opcaoDois == 2:
        listar_cadastro_usuarios()
      else:
        return
    elif opcao_menu == 3:
      print("***REALIZAR BUSCA***\n")
      opcaoTres = int(input("Deseja realizar buscas por: \n1-Veículos \n2-Clientes \n3-Sair\n"))
      if opcaoTres == 1:
        veiculo_estado()
      elif opcaoTres == 2:
        visualizar_cliente()
      else:
        return
    elif opcao_menu == 4:
      print("***ALUGUEL E DEVOLUÇÃO DE VEÍCULO***\n")
      opcaoQuatro = int(input("Deseja realizar qual ação? \n1-Alugar um veículo \n2-Devolver um veículo n\3-Sair\n"))
      if opcaoQuatro == 1:
        alugar_veiculo()
      elif opcaoQuatro == 2:
        devolucao_carro()
      else:
        return
    else:
      print("\n\n*****SESSÃO ENCERRADA*****")
      break
```
