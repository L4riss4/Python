```python
#Criando uma base de dados para uso no sistema
import csv
import os

base_dados = [
    ['PLACA', 'MARCA', 'MODELO', 'ANO', 'COR', 'CLASSIFICACAO'],
    ['ABC-1234', 'Ford', 'Fiesta', 2018, 'Preto', 'Semi-novo'],
    ['DEF-5678', 'Chevrolet', 'Onix', 2020, 'Branco', 'Novo'],
    ['GHI-9101', 'Volkswagen', 'Golf', 2015, 'Cinza', 'Usado'],
    ['JKL-1213', 'Honda', 'Civic', 2019, 'Vermelho', 'Semi-novo'],
    ['MNO-1415', 'Toyota', 'Corolla', 2021, 'Preto', 'Novo'],
    ['PQR-1617', 'Hyundai', 'HB20', 2017, 'Azul', 'Usado'],
    ['STU-1819', 'Nissan', 'Sentra', 2016, 'Prata', 'Usado'],
    ['VWX-2021', 'Kia', 'Sportage', 2018, 'Branco', 'Semi-novo'],
    ['YZA-2223', 'BMW', '320i', 2019, 'Preto', 'Semi-novo'],
    ['BCD-2425', 'Mercedes-Benz', 'Classe C', 2020, 'Cinza', 'Novo'],
    ['EFG-2627', 'Audi', 'A3', 2015, 'Vermelho', 'Usado'],
    ['HIJ-2829', 'Jeep', 'Renegade', 2021, 'Preto', 'Novo'],
    ['KLM-3031', 'Renault', 'Kwid', 2019, 'Branco', 'Semi-novo'],
    ['NOP-3233', 'Peugeot', '208', 2016, 'Azul', 'Usado'],
    ['QRS-3435', 'Citroën', 'C4 Cactus', 2020, 'Cinza', 'Novo'],
    ['TUV-3637', 'Mitsubishi', 'Lancer', 2017, 'Prata', 'Usado'],
    ['WXY-3839', 'Subaru', 'Impreza', 2018, 'Vermelho', 'Semi-novo'],
    ['ZAB-4041', 'Fiat', 'Argo', 2021, 'Branco', 'Novo'],
    ['CDE-4243', 'Suzuki', 'Swift', 2019, 'Preto', 'Semi-novo'],
    ['FGH-4445', 'Lexus', 'NX 300h', 2020, 'Azul', 'Novo'],
    ['IJK-4647', 'Land Rover', 'Evoque', 2015, 'Cinza', 'Usado'],
    ['LMN-4849', 'Porsche', 'Macan', 2018, 'Preto', 'Semi-novo'],
    ['OPQ-5051', 'Jaguar', 'F-Pace', 2019, 'Branco', 'Semi-novo'],
    ['RST-5253', 'Volvo', 'XC60', 2021, 'Vermelho', 'Novo'],
    ['UVW-5455', 'Tesla', 'Model 3', 2020, 'Prata', 'Novo']
]

#Função exportando como arquivo CSV
def exportar_csv(dados, caminho = False):
  '''
    Exporta uma lista de dados para um arquivo CSV.

    Parametros:
    dados (lista de listas): Lista de listas onde cada sub-lista representa uma linha no CSV.
    caminho (str ou bool): Caminho onde o arquivo CSV será salvo. Se False, o arquivo será salvo no diretório atual.

    Retorna:
    None
  '''
  nome = input('Digite o nome do arquivo: ')
  nome =  nome + '.csv'
  if caminho:
    caminho_arquivo = input('Digite o caminho que o arquivo deverá ser salvo: ')
    caminho_arquivo = os.path.join(caminho_arquivo, nome)
  else:
    caminho_arquivo = os.path.join(os.getcwd(), nome)

  try:
    with open(caminho_arquivo, 'w', newline='', encoding='utf-8') as arquivo:
      escrita = csv.writer(arquivo)
      escrita.writerows(dados)
      print(f'Arquivo {nome} salvo com sucesso!')
      return True
  except Exception as e:
    print(f'Erro ao salvar o arquivo {nome}: {e}')
    return False
```

```python
#Função lendo caminhos de arquivos e retornado dataframe pandas
import pandas as pd

def csv_to_df(caminho):
  '''
    Retorna um dataframe pandas a partir de um arquivo csv.

    No Colab é necessário apenas o nome do arquivo que deseja.

    Parametros:
    caminho (str): Caminho completo para o arquivo CSV que deseja ler.

    Retorna:
    df (pandas.DataFrame): Dataframe pandas contendo os dados do arquivo CSV.
  '''
  #caminho_corrigido = caminho.replace('/', '\\')
  #df = pd.read_csv(caminho_corrigido)
  df = pd.read_csv(caminho, sep=",")
  print(f"Dataframe do arquivo {caminho}")
  print(df)
  return df
```

```python
#Função busca por placa com parâmetro de entrada df do banco de dados e placa do carro com retorno em objeto do tipo JSON
import json

def busca_placa(df, placa):
  '''
    Busca por uma placa específica em um DataFrame.

    Parametros:
    - df (pandas.DataFrame): DataFrame contendo os dados do banco de dados.
    - placa (str): Placa do carro que se deseja buscar no DataFrame.

    Retorna:
    - str: Objeto JSON contendo as informações do carro correspondente à placa.
  '''
  busca = None
  busca = df.loc[df['PLACA'] == placa]

  busca = busca.iloc[0].to_dict()
  busca_json = json.dumps(busca, indent=2)
  print(busca_json)
  return busca_json
```

```python
#Função para salvamento dos dados de busca em um arquivo .json, recebendo como parâmetros de entrada um objeto .json chamado dado e o nome do arquivo a ser salvo.
def salva_json(dado, nome):
  '''
    Salva um objeto JSON em um arquivo no caminho especificado pelo usuário ou no diretório atual.

    Parametros:
    - dado (dict): Dicionário contendo os dados a serem serializados para JSON e salvos no arquivo.
    - nome (str): Nome do arquivo JSON a ser salvo. A extensão '.json' será automaticamente adicionada.

    Retorna:
    - None
  '''
  nome = nome + '.json'
  caminho= os.getcwd()
  caminho = os.path.join(caminho, nome)
  with open(nome, 'w', encoding='utf-8') as arquivo:
    json.dump(dado, arquivo, indent=2)
  print(f"Arquivo {nome} salvo com sucesso em {caminho}!")
```

```python
#Função inclusão de dados no dataframe a partir de um objeto json
def incluir_dados(obj_json, df_bd):
  '''
    Inclui os dados de um objeto JSON em um dataframe pandas existente.

    Parametros:
    - obj_json (str): String contendo um objeto JSON com os dados a serem incluidos.
    - df_bd (pandas.DataFrame): Dataframe pandas representando a base de dados existente.

    Retorna:
    - df_bd_atualizado (pandas.DataFrame): Dataframe pandas atualizado com os dados do objeto JSON incluidos.
  '''
  if obj_json.endswith('.json'):
    obj_df = json.loads(obj_json)
    df_atualizado = pd.json_normalize(obj_df)
    df_bd = pd.concat([df_bd, df_atualizado], ignore_index=True)
  else:
    print("Insira um arquivo do tipo .json")
  return df_bd
```

```python
#Função inclusão de df em banco de dados csv
import sqlite3
def incluir_df(df, caminho):
  '''
    Adiciona um DataFrame a um arquivo CSV existente e a um banco de dados SQLite.
    Se o arquivo CSV ou o banco de dados não existir, eles serão criados.

    Parâmetros:
    df (pandas.DataFrame): O DataFrame a ser adicionado ao arquivo CSV e ao banco de dados.
    caminho_csv (str): O caminho do arquivo CSV onde os dados serão adicionados.

    Funcionalidade:
    - Se o arquivo CSV especificado por `caminho_csv` já existir, os dados do DataFrame `df` serão
      concatenados ao final dos dados existentes no arquivo.
    - Se o arquivo CSV não existir, um novo arquivo será criado contendo apenas os dados do DataFrame `df`.
    - O arquivo resultante será salvo no caminho especificado por `caminho_csv`.
    - Os dados do DataFrame também serão armazenados em um banco de dados SQLite chamado `nova_base_dados.db` no diretório atual.
    - Se a tabela `dados` no banco de dados já existir, ela será substituída pelos novos dados do DataFrame `df`.
    - Se a tabela `dados` não existir, ela será criada.

    Mensagem:
    Uma mensagem será impressa indicando que os dados foram adicionados com sucesso ao arquivo CSV e ao banco de dados SQLite, juntamente com os caminhos absolutos dos arquivos salvos.

    Retorna:
    - None
  '''
  if os.path.isfile(caminho):
    df_existente = pd.read_csv(caminho)
    df_novo = pd.concat([df_existente, df], ignore_index=True)
  else:
    df_novo = df

  df_novo.to_csv(caminho, index=False, encoding='utf-8')
  arquivo = os.path.abspath(caminho)
  print(f"Dados adicionados com sucesso em: {arquivo}")

  conexao = sqlite3.connect('nova_tabela.db')
  cursor = conexao.cursor()
  cursor.execute("CREATE TABLE IF NOT EXISTS dados (placa TEXT, marca TEXT, modelo TEXT, ano INTEGER, cor TEXT, classificacao TEXT)")
  df_novo.to_sql('dados', conexao, if_exists='replace', index=False)
  conexao.commit()
  conexao.close()
  print("Dados adicionados com sucesso ao banco de dados SQLite!")
```

```python
#Função inclusão de registro df a partir de um arquivo xls
def incluir_registro(dataframe, arquivo_xls):
  '''
    Inclui os dados de um DataFrame pandas em um arquivo Excel existente (.xls).

    Parâmetros:
    - dataframe (pandas.DataFrame): DataFrame pandas contendo os dados a serem incluídos.
    - arquivo_xls (str): Caminho do arquivo Excel (.xls) onde os dados serão adicionados.

    Funcionalidade:
    - Se o arquivo especificado por `arquivo_xls` tiver a extensão '.xls', os dados do DataFrame
      `dataframe` serão concatenados aos dados existentes no arquivo.
    - Se o arquivo não tiver a extensão '.xls', será exibida uma mensagem de erro solicitando um
      arquivo do tipo correto.
    - O arquivo resultante será salvo no caminho especificado por `arquivo_xls`.

    Retorna:
    - DataFrame: DataFrame atualizado se o arquivo .xls foi carregado corretamente.
    - False: Se houve algum problema ao carregar o arquivo .xls.
  '''
  if arquivo_xls.endswith('.xls'):
    try:
      df_existente = pd.read_excel(arquivo_xls)
      df_novo = pd.concat([df_existente, dataframe], ignore_index=True)
      df_novo.to_excel(arquivo_xls, index=False, encoding='utf-8')
      arquivo_xls = os.path.abspath(arquivo_xls)
      print(f"Dados adicionados com sucesso em: {arquivo_xls}")
      return df_novo
    except Exception as e:
      print(f"Erro ao carregar ou manipular o arquivo {arquivo_xls}: {e}")
  else:
    print("Insira um arquivo do tipo .xls")
    return False
```

```python
#Função busca por carros por classificação
def busca_classificacao(df, classificacao):
  '''
  Filtra um DataFrame com base na classificação fornecida e salva os resultados em um arquivo Excel.

    Parâmetros:
    df (pd.DataFrame): DataFrame contendo os dados com uma coluna 'CLASSIFICACAO'.
    classificacao (str): Classificação a ser buscada ('novo', 'semi-novo' ou 'usado').

    Retorna:
    None: A função salva o resultado da busca em um arquivo Excel e imprime o caminho completo do arquivo.

    O arquivo será salvo no diretório de trabalho atual com o nome 'busca_classificacao.xlsx'.

    Se a classificação fornecida for inválida, uma mensagem de erro será impressa e a função não realizará nenhuma operação de salvamento.
  '''
  classificacao = classificacao.lower()
  if classificacao == 'novo':
    busca = df.loc[df['CLASSIFICACAO'].str.lower() == 'novo']
  elif classificacao == 'semi-novo':
    busca = df.loc[df['CLASSIFICACAO'].str.lower() == 'semi-novo']
  elif classificacao == 'usado':
    busca = df.loc[df['CLASSIFICACAO'].str.lower() == 'usado']
  else:
    print('Classificação inválida')
    return

  nome_arquivo = 'busca_classificacao' + '.xlsx'
  endereco_salvamento = os.path.join(os.getcwd(), nome_arquivo)
  busca.to_excel(endereco_salvamento, index=False)
  arquivo = os.path.abspath(endereco_salvamento)
  print(f"Dados adicionados com sucesso em: {arquivo}")
```

```python
#Função para gerar um dataframe pandas a partir de um banco de dados SQL
import sqlalchemy

def sql_to_df(url, query):
  '''
    Função para executar uma consulta SQL em um banco de dados e retornar um dataframe pandas com os resultados.

    Parâmetros:
    url : str
        URL de conexão com o banco de dados no formato suportado pelo SQLAlchemy, por exemplo:
        'sqlite:///caminho_para_o_seu_banco.db' para SQLite,
        'mysql://usuario:senha@host/nome_do_banco' para MySQL, etc.

    query : str
        Consulta SQL que será executada no banco de dados.

    Retorna:
    pandas.DataFrame or bool
        Um dataframe pandas contendo os resultados da consulta SQL se a conexão e consulta forem bem-sucedidas.
        False se houve algum problema ao conectar ou executar a consulta SQL.
  '''
  try:
    caminho = sqlalchemy.create_engine(url)
    conn = caminho.connect()
    query_texto = sqlalchemy.text(query)
    consulta = conn.execute(query_texto)
    resultado = consulta.fetchall()
    conn.close()

    df_resultado = pd.DataFrame(resultado, columns=consulta.keys())
    return df_resultado
  except Exception as e:
    print(f"Erro ao conectar ou executar a consulta SQL: {e}")
    return False
```

```python
#Função para criar uma base de dados SQL a partir de um df pandas, parametros de entrada são o df com a base de dados e o nome da base de dados SQL que será gerada
def df_to_sql(df, nome_bd):
  '''
  Cria uma base de dados SQL a partir de um dataframe pandas.

    Parâmetros:
    df : pandas.DataFrame
        O dataframe pandas que contém os dados a serem armazenados no banco de dados.

    db_name : str
        O nome da base de dados SQL que será criada.

    Retorna:
    sqlalchemy.engine.base.Engine
        Um objeto Engine do SQLAlchemy que representa a conexão com a base de dados criada.
  '''
  caminho = sqlalchemy.create_engine('sqlite:///bd_dados.db')
  df.to_sql(nome_bd, con=caminho, if_exists='replace', index=False)
  caminho.dispose()
  print("Base de dados criada com sucesso!")
```

```python
#Listagem de tabelas disponiveis para consulta
def listar_tabelas(url):
    """
    Lista as tabelas disponíveis no banco de dados SQLite.

    Args:
        url (str): A URL de conexão para o banco de dados.

    Returns:
        pandas.DataFrame: Um DataFrame contendo os nomes das tabelas.
    """
    engine = sqlalchemy.create_engine(url)
    query = "SELECT name FROM sqlite_master WHERE type='table';"
    tabelas_df = pd.read_sql_query(query, engine)
    engine.dispose()
    return tabelas_df
```

###Webscraping com expressões regulares e beautifulsoup

```python
#Fazendo requisição de site e apresentando conteúdo de forma ordenada
import requests
from bs4 import BeautifulSoup

url = 'https://eventosti.dev/'
response = requests.get(url)

if response.status_code == 200:
  soup = BeautifulSoup(response.content, 'html.parser')
  print(soup.prettify())
else:
  print(f'Erro ao acessar o site: {response.status_code}')
```

```python
#Imprimindo o cabeçalho (head)
if response.status_code == 200:
  soup = BeautifulSoup(response.content, 'html.parser')
  head = soup.find('head')
  if head:
   print(head.prettify())
  else:
    print('Não há head no site')
else:
  print(f'Erro ao acessar o site: {response.status_code}')
```

```python
#Função para busca de elemento passando-o como parametro
def busca_elemento(url, elemento):
  '''
    Faz uma requisição GET a uma URL fornecida e busca por um elemento HTML específico.

    Parametros:
    - url (str): URL do site que será acessado.
    - elemento (str): Tag HTML que será buscada, por exemplo: 'head', 'title', 'div', etc.

    Retorna:
    - str: Retorna o conteúdo do elemento encontrado formatado de forma ordenada (prettify()),
           ou False se o elemento não for encontrado.
  '''
  try:
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')
    elemento = soup.find(elemento)
    if elemento:
      return elemento.prettify()
    else:
      print('Elemento não encontrado')
      return False
  except requests.exceptions.RequestException as e:
    print(f'Erro ao acessar a página: {e}')
    return False
```
