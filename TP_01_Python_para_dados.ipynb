#Cadastro de alunos utilizando lista e dicionário

##**Utilizando apenas listas**



```python
#inicialização
db_alunos = []
id = 0

#Função de cadastro de aluno
def cadastro_alunos():
  global id
  nota_um = 0
  nota_dois = 0
  nota_tres = 0

  nome_aluno = input("Digite o nome completo do aluno: ").title()
  insere_nota = int(input("Deseja inserir as notas do aluno? \n1-Sim \n2-Não \n"))

  if insere_nota == 1:
    nota_um = float(input("Digite a nota 1: "))
    nota_dois = float(input("Digite a nota 2: "))
    nota_tres = float(input("Digite a nota 3: "))
  else:
    print("Notas não foram inseridas")

  id += 1
  media = (nota_um + nota_dois + nota_tres)/3
  db_alunos.append((nome_aluno, nota_um, nota_dois, nota_tres, media, id))

  print(f"\n---Aluno cadastrado--- \n\nNome:{nome_aluno} \nCódigo do aluno: {id} \n\nNotas: \n  01: {nota_um} \n  02: {nota_dois} \n  03: {nota_tres} \nMÉDIA: {media}")

#Função de atualização de notas
def atualizacao_notas():
  print("-----ALUNOS CADASTRADOS-----")
  print("Número - Nome - Código aluno \n")
  for i, alunos in enumerate(db_alunos):
    print(f"{i} - {alunos [0]} - 0{alunos[5]}")

  aluno_atualizar = int(input("Digite o número do aluno que deseja alterar a nota: "))

  if aluno_atualizar < len(db_alunos):
    aluno = db_alunos[aluno_atualizar]

    print("\nDADOS ATUAIS")
    print(f"Aluno (a): {aluno[0]} \nNota 01: {aluno[1]} \nNota 02: {aluno[2]} \nNota 03: {aluno[3]} \nMédia: {aluno[4]}")
    nota_atualizar = int(input("\nQual nota deseja atualizar: \n  1-Nota 01 \n  2-Nota 02 \n  3-Nota 03 \n"))

    if nota_atualizar == 1:
      nota_um_atualizada = float(input("Digite a nova nota: "))
      media_atualizada = (nota_um_atualizada + aluno[2] + aluno[3])/3
      media_formatada = round(media_atualizada, 2)
      db_atualizado = (aluno[0], nota_um_atualizada, aluno[2], aluno[3], media_formatada, aluno[5])
    elif nota_atualizar == 2:
      nota_dois_atualizada = float(input("Digite a nova nota: "))
      media_atualizada = (aluno[1] + nota_dois_atualizada + aluno[3])/3
      media_formatada = round(media_atualizada, 2)
      db_atualizado = (aluno[0], aluno[1], nota_dois_atualizada, aluno[3], media_formatada, aluno[5])
    elif nota_atualizar == 3:
      nota_tres_atualizada = float(input("Digite a nova nota: "))
      media_atualizada = (aluno[1] + aluno[2] + nota_tres_atualizada)/3
      media_formatada = round(media_atualizada, 2)
      db_atualizado = (aluno[0], aluno[1], aluno[2], nota_tres_atualizada, media_formatada, aluno[5])
    else:
      print("Opção escolhida inválida")
      return
    db_alunos[aluno_atualizar] = db_atualizado

    print(f"\nNotas do aluno {aluno [0]} atualizado com sucesso! \nConfira: \n  Nota 01: {db_alunos[aluno_atualizar][1]} \n  Nota 02: {db_alunos[aluno_atualizar][2]} \n  Nota 03: {db_alunos[aluno_atualizar][3]} \n\n    Média: {db_alunos[aluno_atualizar][4]}")

#Função de consulta de aprovados/reprovados
def aprovacao (db, aprovado = True):
  alunos_selecionados = []

  for aluno in db:
    nome_aluno = aluno[0]
    media_aluno = aluno[4]

    if aprovado:
      if media_aluno >= 7:
        status = "Aprovado"
        alunos_selecionados.append((nome_aluno, media_aluno, status))
    else:
      if media_aluno < 7:
        status = "Reprovado"
        alunos_selecionados.append((nome_aluno, media_aluno, status))
  print("******LISTAGEM DE ALUNOS******")
  for aluno, media, aprovacao in alunos_selecionados:
    print(f"\nAluno: {aluno} -------- Média: {media} ------- Status: {aprovacao}")

#Função de consulta por aluno
def listagem_alunos():
  tipo_busca = int(input("Qual o parâmetro de busca? \n1-Nome completo \n2-Código aluno \n"))

  if tipo_busca == 1:
    nome_busca = input("Digite o nome completo do aluno: ").title()

    for aluno in db_alunos:
      if aluno[0] == nome_busca:
        print("Aluno encontrado!")
        print(f"\n---DADOS---  \n\nID: {aluno[5]} \nNome: {aluno[0]} \nNotas \n 01: {aluno[1]} \n 02: {aluno[2]} \n 03: {aluno[3]} \nMÉDIA: {aluno[4]}")
  elif tipo_busca == 2:
    codigo_busca = int(input("informe o código do aluno: "))

    for aluno in db_alunos:
      if aluno[5] == codigo_busca:
        print("Aluno encontrado!")
        print(f"\n---DADOS---  \n\nID: {aluno[5]} \nNome: {aluno[0]} \nNotas \n 01: {aluno[1]} \n 02: {aluno[2]} \n 03: {aluno[3]} \nMÉDIA: {aluno[4]}")

  else:
    print("Digite uma opção válida!")

#Função de impressão de boletim
def boletim():
  print("Deseja gerar boletim de qual aluno?")
  boletim_aluno = int(input("INFORME O CÓDIGO DO ALUNO: "))

  for i in db_alunos:
    if i[5] == boletim_aluno:
      if i[4] >= 7:
        status = "APROVADO"
      else:
        status = "REPROVADO"
      print("\n\n******BOLETIM******")
      print(f"\nNome: {i[0]} ------------ Código identificação: {i[5]}")
      print("\nNOTAS")
      print(f"\n    -Nota 01: {i[1]}")
      print(f"\n    -Nota 02: {i[2]}")
      print(f"\n    -Nota 03: {i[3]}")
      print(f"\nMÉDIA: {i[4]}")
      print(f"\n\nStatus final --------------- {status}")

#Função de inicialização do menu
def menu():
  while True:
    print("\n __________________________________________")
    print("\n   *SISTEMA DE GERENCIAMENTO DE ALUNOS*")
    print("\n1. Cadastrar aluno")
    print("\n2. Editar notas de aluno")
    print("\n3. Listar alunos")
    print("\n4. Gerar boletim")
    print("\n5. Sair")

    opcao = int(input("\nInforme a opção desejada: "))

    if opcao == 1:
      cadastro_alunos()
    elif opcao == 2:
      atualizacao_notas()
    elif opcao == 3:
      tipo_listagem = int(input("\n .Qual tipo de listagem realizar? \n 1-Status de aprovação \n 2-Busca por aluno \n"))
      if tipo_listagem == 1:
        tipo_status = int(input("\n1-Listagem de aprovados \n2-Listagem de reprovados \n"))
        if tipo_status == 1:
          listagem = aprovacao(db_alunos, aprovado = True)
          print(listagem)
        elif tipo_status == 2:
          listagem = aprovacao(db_alunos, aprovado = False)
          print(listagem)
        else:
          print("\nOpção inválida")
      elif tipo_listagem == 2:
        listagem_alunos()
      else:
        print("\nOpção inválida")
    elif opcao == 4:
      boletim()
    elif opcao == 5:
      break
    else:
      print("\n>>>>>>>>INFORME UMA OPÇÃO VÁLIDA<<<<<<<<<")
```

```python
#Inicializando sistema
menu()
```

##**Utilizando apenas dicionários**

```python
#Inicialização

alunos_db = {}
id = 0

#Função de cadastro de alunos
def registro():
  global id
  primeira_nota = 0
  segunda_nota = 0
  terceira_nota = 0
  id += 1

  full_name = input("Digite o nome completo do aluno: ").title()
  insert = int(input("Deseja inserir as notas dos alunos? \n1-Sim \n2-Não \n"))

  if insert == 1:
    primeira_nota = float(input("Digite a nota um: "))
    segunda_nota = float(input("Digite a nota dois: "))
    terceira_nota = float(input("Digite a nota três: "))
  elif insert ==2:
    print("Notas não foram inseridas")
  else:
    print("\nOpção Inválida!")

  gpa = round((primeira_nota + segunda_nota + terceira_nota) / 3, 2)

  alunos_db[id] = {
      'name': full_name,
      'primeira_nota': primeira_nota,
      'segunda_nota': segunda_nota,
      'terceira_nota': terceira_nota,
      'gpa': gpa
  }

  print(f"\n---ALUNO CADASTRADO--- \nNOME: {alunos_db[id]['name']} \nNOTAS \n .Nota 01: {alunos_db[id]['primeira_nota']} \n .Nota 02: {alunos_db[id]['segunda_nota']} \n .Nota 03: {alunos_db[id]['terceira_nota']} \n\nMÉDIA: {alunos_db[id]['gpa']}")

#Função atualização de notas
def update():
  print("-----ALUNOS CADASTRADOS-----")
  print("\n\nID do aluno ------- NOME \n")

  for aluno_id, aluno in alunos_db.items():
    print(f"{aluno_id} ------- {aluno['name']}")

  student_id = int(input("\nInforme o ID do aluno que deseja alterar as notas: "))

  if student_id in alunos_db:
    aluno = alunos_db[student_id]

    print("\nDADOS ATUAIS")
    print(f"Aluno (a): {aluno['name']} \nNota 01: {aluno['primeira_nota']} \nNota 02: {aluno['segunda_nota']} \nNota 03: {aluno['terceira_nota']} \nMédia: {aluno['gpa']}")

    opcoes_notas = {
        1: 'primeira_nota',
        2: 'segunda_nota',
        3: 'terceira_nota'
    }

    grade_change = int(input("Informe qual nota deseja alterar: \n.1-Nota 01 \n.2-Nota 02 \n.3-Nota 03 \n"))

    if grade_change in opcoes_notas:
      nota_atualizada = float(input("Informe a nova nota: "))
      aluno[opcoes_notas[grade_change]] = nota_atualizada

      gpa_atualizada = (aluno['primeira_nota'] + aluno['segunda_nota'] + aluno['terceira_nota']) / 3
      aluno['gpa'] = round(gpa_atualizada, 2)

    print(f"\n>>>>>>DADOS ATUALIZADOS COM SUCESSO<<<<<<")
    print(f"\nAluno: {aluno['name']} \nId: {aluno_id} \nMédia: {aluno['gpa']}")

#Função de consulta de aprovados/reprovados
def aprovar_reprovar():
  opcao = int(input("\nBuscar por: \n1-Aprovados \n2-Reprovados \n"))

  if opcao == 1:
    aprovado = True
  elif opcao == 2:
    aprovado = False
  else:
    print("\nOpção inválida")

  lista_selecionados = []

  for student_id, student in alunos_db.items():
    media_aluno = student['gpa']
    nome_aluno = student['name']
    if aprovado:
      if media_aluno >= 7:
        status_aprovacao = "Aprovado(a)"
        lista_selecionados.append((student_id, nome_aluno, media_aluno, status_aprovacao))
    else:
      if media_aluno < 7:
        status_aprovacao = "Reprovado(a)"
        lista_selecionados.append((student_id, nome_aluno, media_aluno, status_aprovacao))

  print("******LISTAGEM DE ALUNOS******")
  for id, aluno, media, aprovacao in lista_selecionados:
    print(f"\n{id} ------ Aluno: {aluno} -------- Média: {media} ------- Status: {aprovacao}")

#Função consulta por aluno
def consultar_alunos():
  modo_busca = input("\nQual tipo de busca deseja realizar? \n1-Nome completo \n2-ID aluno \n")

  if modo_busca == '1':
    nome_completo = input("\nDigite o nome completo: ").title()
    for student in alunos_db.values():
      if student['name'] == nome_completo:
        print("\nALUNO ENCONTRADO")
        print(f"\nNome: {student['name']} \nNOTAS: \n -Nota 01: {student['primeira_nota']} \n -Nota 02: {student['segunda_nota']} \n -Nota 03: {student['terceira_nota']}")
      else:
        print("\nALUNO NÃO ENCONTRADO")
  elif modo_busca == '2':
    id_aluno = int(input("\nInforme o ID do aluno: "))
    student = alunos_db.get(id_aluno)
    if student:
      print("\nALUNO ENCONTRADO")
      print(f"\nNome: {student['name']} \nNOTAS: \n -Nota 01: {student['primeira_nota']} \n -Nota 02: {student['segunda_nota']} \n -Nota 03: {student['terceira_nota']}")
    else:
      print("\nALUNO NÃO ENCONTRADO")
  else:
    print("\n\nINSIRA UM OPÇÃO VÁLIDA")

#Função de impressão de boletim
def gerar_boletim():
  print("Deseja gerar boletim de qual aluno?")
  id_aluno_boletim= int(input("INFORME O CÓDIGO DO ALUNO: "))

  for student_id, student in alunos_db.items():
    student = alunos_db.get(id_aluno_boletim)
    if student['gpa'] >= 7:
      status = "APROVADO"
    else:
      status = "REPROVADO"
  print("\n\n******BOLETIM******")
  print(f"\nNome: {student['name']} ------------ Código identificação: {student_id}")
  print("\nNOTAS")
  print(f"\n    -Nota 01: {student['primeira_nota']}")
  print(f"\n    -Nota 02: {student['segunda_nota']}")
  print(f"\n    -Nota 03: {student['terceira_nota']}")
  print(f"\nMÉDIA: {student['gpa']}")
  print(f"\n\nStatus final --------------- {status}")

#Função sair do programa
def sair():
  print("\n\n\nPROGRAMA ENCERRADO")

#Função de inicialização do menu
def inicializar_menu():
  menu_opcoes = {
        1: registro,
        2: update,
        3: aprovar_reprovar,
        4: consultar_alunos,
        5: gerar_boletim,
        6: sair
    }

  while True:
      print("\n __________________________________________")
      print("\n   *SISTEMA DE GERENCIAMENTO DE ALUNOS*")
      print("\n1. Cadastrar aluno")
      print("2. Editar notas de aluno")
      print("3. Listar por status de aprovação")
      print("4. Listar alunos")
      print("5. Gerar boletim")
      print("6. Sair")

      try:
          opcao = int(input("\nInforme a opção desejada: "))
          if opcao in menu_opcoes:
              if opcao == 6:
                    menu_opcoes[opcao]()
                    break
              else:
                  menu_opcoes[opcao]()
          else:
              print("\n>>>>>>>>INFORME UMA OPÇÃO VÁLIDA<<<<<<<<<")
      except ValueError:
          print("\nOpção inválida. Por favor, informe um número válido.")
```

```python
#Inicializando o sistema
inicializar_menu()
```
